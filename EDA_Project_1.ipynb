#exploratory data analysis

```python
#ploting a bar chart for gender and its count

ax=sns.countplot(x="Gender",data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# ploting a bar chart gender vs toatalamount
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

sns.barplot(x="Gender",y="Amount",data=sales_gen)# barplot
```

```python
# create count vs age plot

ax=sns.countplot(data=df,x="Age Group")
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=df)# barplot
```

```python
#create count plot for state

ax=sns.countplot(x="State",data=df)
for bars in ax.containers:
  ax.bar_label(bars)
  sns.set(rc={'figure.figsize':(15,5)})
```

```python
#create total number of orders top 10 stat
w=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y="Orders",data=w)# barplot
```

```python
df.info()
```

```python
#create total Amount of orders top 10 state
w=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x="State",y="Amount",data=w)# barplot
```

```python
#total amount sales vs state
d=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='State',y='Amount',data=d)
```

```python
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(20,10)})


ax = sns.countplot(x='Product_Category', data=df, order = df['Product_Category'].value_counts().index)

for bar in ax.containers:
  ax.bar_label(bar)

```

```python
A=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum()
ax=sns.barplot(x='Marital_Status', y='Amount', hue='Gender', data=A)
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
#occupation vs amount barplot
d=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Occupation',y='Amount',hue="Amount",data=d)
for bar in ax.containers:
  ax.bar_label(bar)
sns.set(rc={'figure.figsize':(15,5)})
```

```python
#product vs amount
d=df.groupby(['Product_Category'],as_index=False)['Amount'].sum()
ax=sns.barplot(x='Product_Category',y='Amount',hue="Amount",color="red",data=d)
for bar in ax.containers:
  ax.bar_label(bar)
sns.set(rc={'figure.figsize':(25,5)})# width and height of plot
```

```python
#product_id vs orders
d=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State',y='Amount',data=d)

```
