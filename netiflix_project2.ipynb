```python
'''
Exploratory Data Analysis (EDA) for the Netflix dataset is the process of examining and visualizing Netflix
data to understand its structure, discover patterns such as the number of movies and TV shows, popular genres,
top countries producing content, and trends over the years.
'''
```

```python
import numpy as np #
import pandas as pd  #panadas
import matplotlib.pyplot as plt   #visualization data
%matplotlib inline
import seaborn as sns  #seaborn
```

```python
df=pd.read_csv("/content/netflix_titles_2021.csv",encoding="unicode_escape")
```

```python
df['director'] = df['director'].fillna("xyg")
df['cast'] = df['cast'].fillna("ABCD")
df['country'] = df['country'].fillna("xyg")
df['date_added'] = df['date_added'].fillna("ABCD")
df['rating'] = df['rating'].fillna("MMM")
df['duration'] = df['duration'].fillna("BB")
df
```

```python
pd.isnull(df).sum()
```

```python
df.isnull().sum() # show null is present or not
```

```python
#plotting a bar chart for TV SHOW TYPEA
ax=sns.countplot(x='type',data=df)
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
df.head()
```

```python
#type  vs releas_year
d=df.groupby(['type'],as_index=False)['release_year'].sum()
ax=sns.barplot(x='type',y='release_year',data=d)
for bar in ax.containers:
  ax.bar_label(bar)
  sns.set(rc={'figure.figsize':(25,5)})
```

```python
#number of titale vs releas year
year_counts = df['release_year'].value_counts()

sns.barplot(x=year_counts.index, y=year_counts.values)
plt.xticks(rotation=90)
plt.title("Number of Titles per Release Year")
plt.show()
sns.set(rc={'figure.figsize':(25,5)})
```

### Top 10 Countries by Content

```python
top = df['country'].value_counts().head(10)

sns.barplot(x=top.index, y=top.values, color="red")
plt.xticks(rotation=45)
plt.title("Top 10 Countries by Netflix Content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()


```

### Top 10 Genres (listed_in) by Content

```python
# Since 'listed_in' can contain multiple genres, we need to split and count them
top_genres = df['listed_in'].str.split(', ').explode().value_counts().head(10)

sns.barplot(x=top_genres.index, y=top_genres.values)
plt.xticks(rotation=45)
plt.title("Top 10 Genres on Netflix")
plt.show()
```

```python
# TOP 10 COUNTRIES WITH MOST NETFLIX CONTENT
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
#Top 10 Directors
df['director'].value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Directors on Netflix")
plt.show()
```

```python
#Rating Distribution
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title("Distribution of Content Ratings")
plt.show()
```

```python
#. Most Common Genres
df['listed_in'].str.split(',', expand=True).stack().value_counts().head(10).plot(kind='bar')
plt.title("Top Genres on Netflix")
plt.show()
```

```python
#Content Added Per Year
df['release_year'].value_counts().sort_index().plot(kind='bar')
plt.title("Netflix Content Released Per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```
